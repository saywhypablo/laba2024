4. Сколько мужчин / женщин находилось на борту?

```python
male = titanic_df['Sex'] == 'male'
female = titanic_df['Sex'] == 'female'
print("There were", male.sum(), "men and", female.sum(), "women on the board.")
```

4. Выведите распределение переменной Pclass (социально-экономический статус) и это же распределение, только для мужчин / женщин по отдельности. Сколько было мужчин 2-го класса

```python
pclass = titanic_df['Pclass']
print("Pclass variable distribution:\n", pclass.value_counts())
print("\nPclass variable distribution grouped by male sex:\n", pclass[male].value_counts())
print("\nPclass variable distribution grouped by female sex:\n", pclass[female].value_counts())
count = pclass[male].value_counts().loc[2]
print("\nCount of men in the 2nd class:", count)
```

5. Каковы медиана и среднеквадрати́чное отклонение платежей (Fare)? Округлите до 2 десятичных знаков.

```python
median = round(titanic_df['Fare'].median(), 2)
std = round(titanic_df['Fare'].std(), 2)
print("Median is", median)
print("Standart deviation is", std)
```

6. Правда ли, что люди моложе 30 лет выживали чаще, чем люди старше 60 лет? Каковы доли выживших в обеих группах?

```python
younger = titanic_df[titanic_df['Age'] < 30]
older = titanic_df[titanic_df['Age'] > 60]
ysurvived = round(younger['Survived'].mean(), 2)
osurvived = round(older['Survived'].mean(), 2)
print("Survived percentage for people younger than 30 y.o. is", ysurvived)
print("Survived percentage for people older than 60 y.o. is", osurvived)
print("Given statement is true.") if ysurvived > osurvived else print("Given statement is false.")
```

7. Правда ли, что женщины выживали чаще мужчин? Каковы доли выживших в обеих группах?

```python
male = titanic_df[titanic_df['Sex'] == "male"]
female = titanic_df[titanic_df['Sex'] == "female"]
msurvived = round(male['Survived'].mean(), 2)
fsurvived = round(female['Survived'].mean(), 2)
print("Survived percentage for men is", msurvived)
print("Survived percentage for women is", fsurvived)
print("Given statement is true.") if fsurvived > msurvived else print("Given statement is false.")
```

8.  Найдите самое популярное имя среди пассажиров Титаника мужского пола?

```python
name = male['Name'].str.split(" ").str[2]
print(f"{name.mode().iloc[0]} is the most common name among men on the Titanic.")
```



9. Как отличается средний возраст мужчин / женщин в зависимости от класса обслуживания? Выберите верные утверждения:

    1) В среднем мужчины 1-го класса старше.

    2) В среднем женщины 1-го класса старше 40 лет.

    3) Мужчины всех классов в среднем старше женщин того же класса.

    4) В среднем люди в 1 классе старше, чем во 2-ом, а те старше представителей 3-го класса.



```python
mean_age = titanic_df.groupby(['Pclass', 'Sex'])['Age'].mean()
print(mean_age)

print("\nAs follows, the first and the third statements are true.")
```

10. Сравните возраст у спасенных и у погибших пасажиров. Средний возраст погибших выше, верно?

```python
survived = titanic_df[titanic_df['Survived'] == 1]
dead = titanic_df[titanic_df['Survived'] == 0]
survived_age = round(survived['Age'].mean(), 2)
dead_age = round(dead['Age'].mean(), 2)
print("Mean age for survived is", survived_age)
print("Mean age for dead is", dead_age)
print("Given statement is true.") if dead_age > survived_age else print("Given statement is false.")
```
